```python
a_df = df[['AAPL']].dropna()
a_df
```

```python
window = 20

a_df['min'] = a_df['AAPL'].rolling(window=window).min()
a_df['mean'] = a_df['AAPL'].rolling(window=window).mean()
a_df['std'] = a_df['AAPL'].rolling(window=window).std()
a_df['median'] = a_df['AAPL'].rolling(window=window).median()
a_df['max'] = a_df['AAPL'].rolling(window=window).max()

a_df.dropna()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

ax = a_df[['min', 'mean','median', 'max']].iloc[-252:].plot(figsize=(12, 6), style=['g--', 'r--', 'm--', 'g--'], lw=0.8)
a_df['AAPL'].iloc[-252:].plot(ax=ax)
plt.title("AppleÏùò 20Ïùº Ïù¥ÎèôÌèâÍ∑† Í∞ÄÍ≤© Î≥ÄÎèôÎ•†")
plt.show()
```

```python
a_df['SMA1'] = a_df['AAPL'].rolling(window=21).mean() #short-term
a_df['SMA2'] = a_df['AAPL'].rolling(window=252).mean() #long-term
a_df[['AAPL', 'SMA1', 'SMA2']].tail()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

a_df[['AAPL', 'SMA1', 'SMA2']].plot(figsize=(12, 6))
plt.title("Appl Îã®Ïàú Ïù¥Îèô ÌèâÍ∑† - Ïû•Í∏∞ Î∞è Îã®Í∏∞")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

a_df.dropna(inplace=True)

a_df['positions'] = np.where(a_df['SMA1'] > a_df['SMA2'], 1, -1)  # 1: buy , -1: sell /

ax = a_df[['AAPL', 'SMA1', 'SMA2', 'positions']].plot(figsize=(15, 6), secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((-0.05, 1))

plt.title("Í∏∞Ïà†Ï†Å Î∂ÑÏÑùÏùÑ Í∏∞Î∞òÏúºÎ°ú Ìïú Apple Í±∞Îûò Ï∞ΩÍµ¨")
plt.show()
```

### üìàARIMA Î™®Îç∏ÏùÑ ÏÇ¨Ïö©Ìïú Î™®Îç∏ ÌèâÍ∞Ä ÏßÑÌñâ

```python
df
```

```python
a_df1 = df.drop(columns=['GLD'])
a_df1
```

```python
a_df1.info()
```

```python
a_df1 = a_df1.dropna()
a_df1 = a_df1['2023-01-03':]
a_df1
```

```python
y_train = a_df1['AAPL'][:int(0.8 * len(a_df1))]
y_test = a_df1['AAPL'][int(0.8 * len(a_df1)):]
```

```python
y_train.plot()
y_test.plot()
```

```python
from pmdarima.arima import ndiffs

kpss_diffs = ndiffs(y_train, alpha=0.05, test='kpss', max_d=10)
adf_diffs = ndiffs(y_train, alpha=0.05, test='adf', max_d=10)
pp_diffs = ndiffs(y_train, alpha=0.05, test='pp', max_d=10)

n_diffs = max(kpss_diffs, adf_diffs, pp_diffs)

print(n_diffs)
```

```python
import pmdarima as pm

model = pm. auto_arima(y=y_train, 
                       start_p=0, max_p=10, 
                       start_q=0, max_q=10, 
                       m=1, stepwise=True, 
                       trace=True)
```

```python
import matplotlib.pyplot as plt

model.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
prediction = model.predict(n_periods=len(y_test))
prediction
```

```python
prediction, conf_int = model.predict(n_periods=1, return_conf_int=True)
print(conf_int)
print(prediction)
```

```python
prediction.tolist()[0]
```

```python
def predict_one_step():
    prediction = model.predict(n_periods=1)
    return prediction.tolist()[0]
```

```python
p_list = []

for data in y_test:
    p = predict_one_step()
    p_list.append(p)

    model.update(data)
```

```python
import pandas as pd

y_predict_df = pd.DataFrame({"test": y_test, "pred": p_list})
y_predict_df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

plt.plot(y_train, label='Train')
plt.plot(y_test, label='Test')
plt.plot(y_predict_df.pred, label='Prediction')
plt.legend()
plt.show()
```

```python
import numpy as np

def MAPE(y_test, y_pred):
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print(f'MAPE (%): {MAPE(y_test, p_list):.4f}')
```

```python
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
import itertools

# changepoint_prior_scale: trendÏùò Î≥ÄÌôîÌïòÎäî ÌÅ¨Í∏∞Î•º Î∞òÏòÅÌïòÎäî Ï†ïÎèÑÏù¥Îã§, 0.05Í∞Ä default
# seasonality_prior_scale: Í≥ÑÏ†àÏÑ±ÏùÑ Î∞òÏòÅÌïòÎäî Îã®ÏúÑÏù¥Îã§.
# seasonality_mode: Í≥ÑÏ†àÏÑ±ÏúºÎ°ú ÎÇòÌÉÄÎÇòÎäî Ìö®Í≥ºÎ•º ÎçîÌï¥ ÎÇòÍ∞àÏßÄ, Í≥±Ìï¥ ÎÇòÍ∞àÏßÄ Ï†ïÌïúÎã§.
search_space = {
    'changepoint_prior_scale': [0.05, 0.1, 0.5, 1.0, 5.0, 10.0],
    'seasonality_prior_scale': [0.05, 0.1, 1.0, 10.0],
    'seasonality_mode': ['additive', 'multiplicative']
}

# itertools.product(): Í∞Å ÏöîÏÜåÎì§Ïùò Î™®Îì† Í≤ΩÏö∞Ïùò Ïàò Ï°∞Ìï©ÏúºÎ°ú ÏÉùÏÑ±
param_combinded = [dict(zip(search_space.keys(), v)) for v in itertools.product(*search_space.values())]

train_len = int(len(pre_a_df) * 0.8)
test_len = int(len(pre_a_df) * 0.2)

train_size = f'{train_len} days'
test_size = f'{test_len} days'
train_df = pre_a_df.iloc[: train_len]
test_df = pre_a_df.iloc[train_len: ]

mapes = []
for param in param_combinded:
    model = Prophet(**param)
    model.fit(train_df)

    # 'threads' ÏòµÏÖòÏùÄ Î©îÎ™®Î¶¨ ÏÇ¨Ïö©ÎüâÏùÄ ÎÇÆÏßÄÎßå CPU Î∞îÏö¥Îìú ÏûëÏóÖÏóêÎäî Ìö®Í≥ºÏ†ÅÏù¥ÏßÄ ÏïäÏùÑ Ïàò ÏûàÎã§.
    # 'dask' ÏòµÏÖòÏùÄ ÎåÄÍ∑úÎ™®Ïùò Îç∞Ïù¥ÌÑ∞Î•º Ï≤òÎ¶¨ÌïòÎäî Îç∞ Ìö®Í≥ºÏ†ÅÏù¥Îã§.
    # 'processes' ÏòµÏÖòÏùÄ Í∞ÅÍ∞ÅÏùò ÏûëÏóÖÏùÑ Î≥ÑÎèÑÏùò ÌîÑÎ°úÏÑ∏Ïä§Î°ú Ïã§ÌñâÌïòÍ∏∞ ÎïåÎ¨∏Ïóê CPU Î∞îÏö¥Îìú ÏûëÏóÖÏóê Ìö®Í≥ºÏ†ÅÏù¥ÏßÄÎßå,
    # Î©îÎ™®Î¶¨ ÏÇ¨Ïö©ÎüâÏù¥ ÎÜíÏùÑ Ïàò ÏûàÎã§.
    cv_df = cross_validation(model, initial=train_size, period='20 days', horizon=test_size, parallel='processes')
    df_p = performance_metrics(cv_df, rolling_window=1)
    mapes.append(df_p['mape'].values[0])

tuning_result = pd.DataFrame(param_combinded)
tuning_result['mape'] = mapes       
```

```python
model = Prophet(changepoint_prior_scale=0.5, 
                seasonality_prior_scale=0.1, 
                seasonality_mode='additive')

model.fit(pre_a_df)
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```
