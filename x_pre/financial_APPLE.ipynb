```python
import yfinance as yf

columns = ['AAPL', 'GLD']

df = yf.download(columns, start='2008-01-02')['Adj Close'].round(4)
df
```

### üìà Î≥ÄÎèôÎ•† log ÏãúÍ∞ÅÌôî

```python
apple_stock_df = df.copy()
apple_stock_df = apple_stock_df.drop(labels='GLD', axis=1)
apple_stock_df
```

```python
y_train = a_df1['AAPL'][:int(0.8 * len(a_df1))]
y_test = a_df1['AAPL'][int(0.8 * len(a_df1)):]
```

```python
y_train.plot()
y_test.plot()
```

```python
from pmdarima.arima import ndiffs

kpss_diffs = ndiffs(y_train, alpha=0.05, test='kpss', max_d=10)
adf_diffs = ndiffs(y_train, alpha=0.05, test='adf', max_d=10)
pp_diffs = ndiffs(y_train, alpha=0.05, test='pp', max_d=10)

n_diffs = max(kpss_diffs, adf_diffs, pp_diffs)

print(n_diffs)
```

```python
import pmdarima as pm

model = pm. auto_arima(y=y_train, 
                       start_p=0, max_p=10, 
                       start_q=0, max_q=10, 
                       m=1, stepwise=True, 
                       trace=True)
```

```python
import matplotlib.pyplot as plt

model.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
prediction = model.predict(n_periods=len(y_test))
prediction
```

```python
prediction, conf_int = model.predict(n_periods=1, return_conf_int=True)
print(conf_int)
print(prediction)
```

```python
prediction.tolist()[0]
```

```python
def predict_one_step():
    prediction = model.predict(n_periods=1)
    return prediction.tolist()[0]
```

```python
p_list = []

for data in y_test:
    p = predict_one_step()
    p_list.append(p)

    model.update(data)
```

```python
import pandas as pd

y_predict_df = pd.DataFrame({"test": y_test, "pred": p_list})
y_predict_df
```

```python
import numpy as np

def MAPE(y_test, y_pred):
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print(f'MAPE (%): {MAPE(y_test, p_list):.4f}')
```

```python
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
import itertools

search_space = {
    'changepoint_prior_scale': [0.05, 0.1, 0.5, 1.0, 5.0, 10.0],
    'seasonality_prior_scale': [0.05, 0.1, 1.0, 10.0],
    'seasonality_mode': ['additive', 'multiplicative']
}

param_combinded = [dict(zip(search_space.keys(), v)) for v in itertools.product(*search_space.values())]

train_len = int(len(pre_a_df) * 0.8)
test_len = int(len(pre_a_df) * 0.2)

train_size = f'{train_len} days'
test_size = f'{test_len} days'
train_df = pre_a_df.iloc[: train_len]
test_df = pre_a_df.iloc[train_len: ]

mapes = []
for param in param_combinded:
    model = Prophet(**param)
    model.fit(train_df)

    cv_df = cross_validation(model, initial=train_size, period='20 days', horizon=test_size, parallel='processes')
    df_p = performance_metrics(cv_df, rolling_window=1)
    mapes.append(df_p['mape'].values[0])

tuning_result = pd.DataFrame(param_combinded)
tuning_result['mape'] = mapes       
```

```python
model = Prophet(changepoint_prior_scale=0.5, 
                seasonality_prior_scale=0.1, 
                seasonality_mode='additive')

model.fit(pre_a_df)
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```
