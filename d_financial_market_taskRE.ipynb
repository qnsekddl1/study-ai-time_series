### 비트 코인 가격 예측  
https://www.blockchain.com/explorer/charts/market-price

- auto_arima를 사용하여 적합한 모델을 찾고, 이를 통해 금융 데이터 포트폴리오를 제작한다.

```python
import pandas as pd
import json

with open('./datasets/b_2010_2024.json') as f:
    json_data = json.load(f)

df = pd.DataFrame(json_data['market-price'])
df
```

```python
from datetime import datetime

def changeDate(milis):
    timestamp = milis / 1000
    converted_time = datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')
    return converted_time
```

```python
df.x = df.x.apply(changeDate)
df
```

```python
df.info()
```

```python
df.set_index('x', inplace=True)
df
```

```python
df.info()
```

```python
df.index = pd.to_datetime(df.index)
df.info()
```

```python
df.isna().sum()
```

```python
pre_df = df.dropna()
pre_df = pre_df['2017-01-01':]
pre_df
```

```python
y_train = pre_df['y'][:int(0.8 * len(pre_df))]
y_test = pre_df['y'][int(0.8 * len(pre_df)):]
```

```python
y_train.plot()
y_test.plot()
```

```python
from pmdarima.arima import ndiffs

kpss_diffs = ndiffs(y_train, alpha=0.05, test='kpss', max_d=10)
adf_diffs = ndiffs(y_train, alpha=0.05, test='adf', max_d=10)
pp_diffs = ndiffs(y_train, alpha=0.05, test='pp', max_d=10)

n_diffs = max(kpss_diffs, adf_diffs, pp_diffs)

print(n_diffs)
```

```python
import pmdarima as pm

model = pm. auto_arima(y=y_train, 
                       start_p=0, max_p=10, 
                       start_q=0, max_q=10, 
                       m=1, stepwise=True, 
                       trace=True)
```

```python
model.fit(y_train)
```

```python
print(model.summary())
```

### Summary 정리
1. Prob(Q)가 0.99로 0.05 이상이기 때문에 서로 독립적이고 동일한 분포에 가까울 것으로 예상됩니다.
2. Prob(H)가 0으로 잔차의 분산이 일정하지 않은 것으로 보입니다.
3. Prob(JB)가 0이기 때문에 일정한 평균과 분산을 따르지 않을 것으로 예상됩니다.
4. Skew가 -0.33이기 때문에 분포가 약간 왼쪽으로 치우쳐 있음을 의미합니다.  
   하지만 이 정도의 Skew는 크게 비대칭적이지 않으며, 데이터가 거의 대칭적으로 분포되어 있다고 불 수 있습니다.
6. Kurtosis가 9.44인 것은 높은 수치이며 데이터들이 중앙에 많이 밀집되어 있다고 볼 수 있습니다.

```python
import matplotlib.pyplot as plt

model.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
prediction = model.predict(n_periods=len(y_test))
prediction
```

```python
prediction, conf_int = model.predict(n_periods=1, return_conf_int=True)
print(conf_int)
print(prediction)
```

```python
prediction.tolist()[0]
```

```python
def predict_one_step():
    prediction = model.predict(n_periods=1)
    return prediction.tolist()[0]
```

```python
p_list = []

for data in y_test:
    p = predict_one_step()
    p_list.append(p)

    model.update(data)
```

```python
y_predict_df = pd.DataFrame({"test": y_test, "pred": p_list})
y_predict_df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

plt.plot(y_train, label='Train')
plt.plot(y_test, label='Test')
plt.plot(y_predict_df.pred, label='Prediction')
plt.legend()
plt.show()
```

```python
import numpy as np

def MAPE(y_test, y_pred):
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print(f'MAPE (%): {MAPE(y_test, p_list):.4f}')
```

```python
window = 10

pre_df['min'] = pre_df['y'].rolling(window=window).min()
pre_df['mean'] = pre_df['y'].rolling(window=window).mean()
pre_df['std'] = pre_df['y'].rolling(window=window).std()
pre_df['median'] = pre_df['y'].rolling(window=window).median()
pre_df['max'] = pre_df['y'].rolling(window=window).max()

pre_df.dropna()
```

```python
import matplotlib.pyplot as plt

ax = pre_df[['min', 'mean', 'max']].iloc[-252:].plot(figsize=(12, 6), style=['g--', 'r--', 'g--'], lw=0.8)
pre_df['y'].iloc[-252:].plot(ax=ax)
plt.title("Bitcoin 10-Day Moving Average Price Movement")
plt.show()
```

```python
# SMA(Simple Moving Average): 일정 기간동안의 가격의 평균을 나타내는 보조지표
# 1달 영업일을 21일로 가정, 1년 영업일을 252일로 가정

pre_df['SMA1'] = pre_df['y'].rolling(window=21).mean() #short-term
pre_df['SMA2'] = pre_df['y'].rolling(window=252).mean() #long-term
pre_df[['y', 'SMA1', 'SMA2']].tail()
```

```python
pre_df[['y', 'SMA1', 'SMA2']].plot(figsize=(12, 6))
plt.title("Bitcoin Simple Moving Average - Long & Short-term")
plt.show()
```

```python
# 아마존 주가 기술 분석
# 골든 크로스, 데드 크로스
pre_df.dropna(inplace=True)

pre_df['positions'] = np.where(pre_df['SMA1'] > pre_df['SMA2'], 1, -1)  # 1: buy , -1: sell /

ax = pre_df[['y', 'SMA1', 'SMA2', 'positions']].plot(figsize=(15, 6), secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((-0.05, 1))

plt.title("Bitcoin Trading Window based on Technical Analysis")
plt.show()
```
