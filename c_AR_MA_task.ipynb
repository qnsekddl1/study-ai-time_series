**처음으로 0에 수렴하는 시차이후에 상관이 크게 줄어드는 지점이 적절한 차분 횟수이다.**  
**하지만 일번적인 기준점이므로 다를 수는 있기 때문에 여러 차분을 시도하여 결정해야 한다.**

```python
import pandas as pd
import json

with open('./datasets/bitcoin_2010_2024.json') as f:
    json_data = json.load(f)

b_df = pd.DataFrame(json_data['market-price'])
b_df
```

```python
from datetime import datetime

def changeDate(milis):
    timestamp = milis / 1000
    converted_time = datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')
    return converted_time
```

```python
b_df.x = b_df.x.apply(changeDate)
b_df
```

```python
b_df.info()
```

```python
b_df.set_index('x', inplace=True)
b_df
```

```python
b_df.info()
```

```python
b_df.index = pd.to_datetime(b_df.index)
b_df.info()
```

```python
b_df.isna().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

# 전체 기간동안의 그래프 확인
plt.plot(b_df['2009-01-03':], c='blue')
plt.title('Full Time Graph')
plt.show()
```

```python
plt.figure(figsize=(15, 5))

# 2017년 부터의 부분 그래프
plt.plot(b_df['2017-01-01':], c='skyblue')
plt.title('Partial Time Graph')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

fig = plot_acf(b_df, lags=20, ax=ax[0])
fig = plot_pacf(b_df, lags=20, ax=ax[1])
```

<!-- PACF 그래프를 보았을 때, 처음으로 0에 수렴 하는 곳이 3차 였고, 그 다음으로 가장 멀리 떨어져있는 곳이 5차 였다.  
<strong>∴ 따라서 ARIMA 모델 사용 시 적절한 차분은 5차 차분이 될듯 하다.</strong> -->
---

```python
b_df.diff().diff().diff().diff().diff().dropna()
```

```python
plt.figure(figsize=(15, 5))

# 2017년 부터의 부분 그래프
plt.plot(b_df['2017-01-01':].diff().dropna(), c='skyblue')
plt.plot(b_df['2017-01-01':].diff().diff().dropna(), c='blue', alpha=0.5)
plt.plot(b_df['2017-01-01':].diff().diff().diff().dropna(), c='red', alpha=0.2)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_pacf(b_df.diff().dropna(), lags=20)
plt.show()
```

```python
plt.figure(figsize=(15, 5))

window_size = 21

# 2017년 부터의 부분 그래프
plt.plot(b_df['2017-01-01':].diff().dropna(), c='skyblue', alpha=0.5)
plt.plot(b_df['2017-01-01':].diff().dropna().rolling(window_size).mean(), c='blue')
plt.show()
```

```python
b_df = b_df['2017-01-01':]
b_df
```

```python
b_df.tail()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import pandas as pd

model = ARIMA(b_df, order=(1, 4, 21))
model_fit = model.fit()

start_index = pd.to_datetime('2023-01-02')
end_index = pd.to_datetime('2024-06-09')
forecast = model_fit.predict(start=start_index, end=end_index)

plt.figure(figsize=(15, 8))
plt.plot(b_df, label='original')
plt.plot(forecast, label='predicted', c='red')
plt.title("Time Series Forecast")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_squared_log_error

# order=(1, 1, 21)
# Mean Squared Error 3834201.7009104197
# Root Mean Squared Error 1958.111769258951
# Mean Squared Error 0.0024821409083510067

# order=(4, 1, 21)
# Mean Squared Error 3675874.224743128
# Root Mean Squared Error 1917.2569532389568
# Mean Squared Error 0.002402070496645408

# order=(1, 4, 21)
# Mean Squared Error 3912598.0063902796
# Root Mean Squared Error 1978.0288183922598
# Mean Squared Error 0.002543464837190769

mse = mean_squared_error(b_df['2023-01-02':'2024-06-09'], forecast)
print("Mean Squared Error", mse)

mse = mean_squared_error(b_df['2023-01-02':'2024-06-09'], forecast)
print("Root Mean Squared Error", mse ** (1/2))

msle = mean_squared_log_error(b_df['2023-01-02':'2024-06-09'], forecast)
print("Mean Squared Error", msle)
```
